```python
import pandas as pd

# Load the datasets
file_path_genetics = 'C:\\Users\\trejan\\Desktop\\Sem 2\\Machine Learning\\model\\genetic_data_detailed.csv'  # Genetics dataset
file_path_nutrition = "C:\\Users\\trejan\\Desktop\\Sem 2\\Machine Learning\\model\\train.csv"  # Update this path with actual location


genetics_df = pd.read_csv(file_path_genetics)
nutrition_df = pd.read_csv(file_path_nutrition)

# Display basic info
def explore_data(df, name):
    print(f'\n{name} Dataset Info:')
    print(df.info())
    print(f'\n{name} Missing Values:')
    print(df.isnull().sum())
    print(f'\n{name} Sample Data:')
    print(df.head())

# Explore both datasets
explore_data(genetics_df, 'Genetics')
explore_data(nutrition_df, 'Nutrition')
```

```python
import pandas as pd

# Define the file path
file_path = r"C:\Users\trejan\Desktop\Sem 2\Machine Learning\GenoneDataset\genes\gene_info"

# Load the data (adjust the delimiter based on your file structure)
# For example, if it's comma-separated, use sep=',', or for tab-separated use sep='\t'
data = pd.read_csv(file_path, sep="\t", header=0)  # Change sep if needed

# Define the genes to filter
genes_of_interest = ['FTO', 'MC4R', 'LEPR', 'PPARG', 'ADIPOQ']

# Filter the data
filtered_data = data[data['Gene Name'].isin(genes_of_interest)]  # Adjust 'Gene' if your column name is different

# Display the filtered data
print(filtered_data)

# Optionally, save the filtered data to a new file
filtered_data.to_csv(r"C:\Users\trejan\Desktop\filtered_genes.csv", index=False)
```
