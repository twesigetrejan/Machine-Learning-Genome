```python
import pandas as pd
from sklearn.preprocessing import StandardScaler, LabelEncoder
from sklearn.model_selection import StratifiedKFold
from sklearn.metrics import classification_report, accuracy_score
from sklearn.ensemble import RandomForestClassifier

# Load datasets
genetic_profiles_path = 'C:\\Users\\trejan\\Desktop\\Sem 2\\Machine Learning\\model\\genetic_profiles.csv'
meal_database_path = 'C:\\Users\\trejan\\Desktop\\Sem 2\\Machine Learning\\model\\train.csv'

genetic_df = pd.read_csv(genetic_profiles_path)
meal_df = pd.read_csv(meal_database_path)

# Merge datasets based on a common identifier (Profile_ID)
combined_df = pd.merge(genetic_df, meal_df, how='inner', left_on='Profile_ID', right_on='ID')

# Drop unnecessary columns
combined_df.drop(columns=['ID'], inplace=True)

# Check for missing values
print(combined_df.isnull().sum())

# Select features and target variable
X = combined_df[['Age', 'BMI', 'Physical_Activity', 'Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g']]
y = combined_df['Diet_Type']  # Assuming 'Diet_Type' is your target variable

# Encode categorical variables
X.loc[:, 'Physical_Activity'] = LabelEncoder().fit_transform(X['Physical_Activity'])
y = LabelEncoder().fit_transform(y)  # Encoding target variable as well

# Normalize continuous features using .loc
scaler = StandardScaler()
X.loc[:, ['Age', 'BMI', 'Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g']] = scaler.fit_transform(X[['Age', 'BMI', 'Energy_kcal', 'Protein_g', 'Fat_g', 'Carb_g']])

# Use Stratified K-Folds for cross-validation
skf = StratifiedKFold(n_splits=5, shuffle=True, random_state=42)
accuracies = []

for train_index, test_index in skf.split(X, y):
    X_train, X_test = X.iloc[train_index], X.iloc[test_index]
    y_train, y_test = y[train_index], y[test_index]

    # Train the model
    model = RandomForestClassifier(random_state=42)
    model.fit(X_train, y_train)

    # Predict and evaluate
    y_pred = model.predict(X_test)
    accuracies.append(accuracy_score(y_test, y_pred))

# Output mean accuracy
mean_accuracy = sum(accuracies) / len(accuracies)
print(f"Mean accuracy across folds: {mean_accuracy:.2f}")

# Optional: Print classification report for the last fold
print("\nClassification report for the last fold:")
print(classification_report(y_test, y_pred))
```
