```python
import pandas as pd

# Load datasets
genetic_df = pd.read_csv("genetic_data_detailed.csv")  # Updated dataset

print(genetic_df)
# Display basic information about the dataset
print(genetic_df.info())

# Show the first few rows
print(genetic_df.head())

# Check for missing values
print(genetic_df.isnull().sum())
```

```python
# Count occurrences of each gene type
gene_type_counts = genetic_df["Gene Type"].value_counts()
import matplotlib.pyplot as plt
import seaborn as sns
# Plot
plt.figure(figsize=(8, 8))
plt.pie(gene_type_counts, labels=gene_type_counts.index, autopct="%1.1f%%", colors=sns.color_palette("pastel"))
plt.title("Distribution of Gene Types")
plt.show()
```

```python
plt.figure(figsize=(10, 5))
sns.heatmap(genetic_df.isnull(), cmap="coolwarm", cbar=False, yticklabels=False)
plt.title("Missing Data Heatmap")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.histplot(genetic_df["Chromosome"], bins=20, kde=True, color="blue")
plt.xlabel("Chromosome")
plt.ylabel("Gene Count")
plt.title("Histogram of Genes Across Chromosomes")
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.boxplot(x=genetic_df["Chromosome"], color="lightblue")
plt.xlabel("Chromosome Number")
plt.title("Box Plot of Chromosome Distribution")
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.countplot(y="Effect of SNP", data=genetic_df, hue="Effect of SNP", legend=False, palette="pastel")
plt.xlabel("Count")
plt.ylabel("Effect of SNP")
plt.title("Distribution of SNP Effects")
plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.countplot(x="Chromosome", data=genetic_df, palette="viridis")

plt.xlabel("Chromosome Number", fontsize=12)
plt.ylabel("Number of Genes", fontsize=12)
plt.title("Gene Distribution Across Chromosomes", fontsize=14)

plt.xticks(rotation=45)  # Rotate x-axis labels if needed
plt.grid(axis="y", linestyle="--", alpha=0.7)
plt.show()
```

```python
plt.figure(figsize=(8, 5))
sns.barplot(x=gene_type_counts.index, y=gene_type_counts.values, palette="pastel")

plt.xlabel("Gene Type", fontsize=12)
plt.ylabel("Count", fontsize=12)
plt.title("Gene Type Distribution", fontsize=14)
plt.grid(axis="y", linestyle="--", alpha=0.7)

plt.show()
```

```python
plt.figure(figsize=(12, 6))
sns.boxplot(x="Chromosome", y="Gene Type", data=genetic_df, palette="pastel")

plt.xlabel("Chromosome", fontsize=12)
plt.ylabel("Gene Type", fontsize=12)
plt.title("Gene Type Distribution Across Chromosomes", fontsize=14)

plt.xticks(rotation=45)
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Choose two meaningful numeric pairs
plt.figure(figsize=(12, 5))

# Scatter plot 1
plt.subplot(1, 2, 1)
sns.scatterplot(x=genetic_df["Chromosome"], y=genetic_df["Gene ID"], alpha=0.6, color="blue")
plt.xlabel("Chromosome")
plt.ylabel("Gene ID")
plt.title("Chromosome vs Gene ID")

# Scatter plot 2
plt.subplot(1, 2, 2)
sns.scatterplot(x=genetic_df["Chromosome"], y=genetic_df["Associated Diseases"], alpha=0.6, color="red")
plt.xlabel("Chromosome")
plt.ylabel("Associated Diseases")
plt.title("Chromosome vs Associated Diseases")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Choose two meaningful numeric pairs
plt.figure(figsize=(12, 5))

# Scatter plot 1
plt.subplot(1, 2, 1)
sns.scatterplot(x=genetic_df["Chromosome"], y=genetic_df["Gene ID"], alpha=0.6, color="blue")
plt.xlabel("Chromosome")
plt.ylabel("Gene ID")
plt.title("Chromosome vs Gene ID")

# Scatter plot 2
plt.subplot(1, 2, 2)
sns.scatterplot(x=genetic_df["Chromosome"], y=genetic_df["Associated Diseases"], alpha=0.6, color="red")
plt.xlabel("Chromosome")
plt.ylabel("Associated Diseases")
plt.title("Chromosome vs Associated Diseases")

plt.tight_layout()  # Adjust layout for better spacing
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

# Define variable pairs
var_pairs = [("Chromosome", "Gene ID"), ("Chromosome", "Associated Diseases"), 
             ("Gene ID", "Associated Diseases")]

# Create subplots dynamically
fig, axes = plt.subplots(1, 3, figsize=(15, 5))  # 1 row, 3 columns

# Loop through variable pairs and create scatter plots
for i, (x_var, y_var) in enumerate(var_pairs):
    sns.scatterplot(x=genetic_df[x_var], y=genetic_df[y_var], alpha=0.6, ax=axes[i])
    axes[i].set_xlabel(x_var)
    axes[i].set_ylabel(y_var)
    axes[i].set_title(f"{x_var} vs {y_var}")

plt.tight_layout()
plt.show()
```

```python
plt.figure(figsize=(10, 6))
sns.swarmplot(x=genetic_df["Chromosome"], y=genetic_df["Gene ID"], hue=genetic_df["Gene Type"], palette="Set2", alpha=0.7)
plt.xlabel("Chromosome")
plt.ylabel("Gene ID")
plt.title("Distribution of Gene Mutations Across Chromosomes")
plt.show()
```
